```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.flipkart.com/search?q=mobile+under+10000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&sort=recency_desc")
c = r.content
c
```

```python
soup=BeautifulSoup(c, "html.parser")
soup
```

```python
all= soup.find_all("div",{"class":"tUxRFH"})
all
```

```python
l =[]
for item in all:
    d={}
    d["phone_name"]=item.find_all("div",{"class":"KzDlHZ"})[0].text
    d["description"]=item.find_all("div",{"class":"_6NESgJ"})[0].text
    d["price"]=item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text
    l.append(d)
l
```

```python
import pandas
df = pandas.DataFrame(l)
df   
```

```python
df.to_csv("webscraping.csv")
```
